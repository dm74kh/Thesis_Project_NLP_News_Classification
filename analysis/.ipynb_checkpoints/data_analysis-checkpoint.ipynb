# Exploratory Data Analysis (EDA) and Text Preprocessing Pipeline of the News Classification Dataset

```python
# Import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from collections import Counter
from scipy.stats import shapiro, normaltest

# Initialize the stopword list and lemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Downloading 'punkt' for tokenizer
nltk.download('punkt')
```

### Loading and previewing the data

```python
df_train = pd.read_csv('data/agn_train.csv') # loading the dataset

df_train.head() # previewing the five first rows
```

```python
df_train.info() # displaying a summary of the dataset with data types
```

### Text Data Cleaning and Preprocessing

```python
# Preprocessing function

def preprocess_text(text):
    text = text.lower() # convert text to lowercase
    
    text = re.sub(r'\b(u\.s\.|us)\b', 'usa', text, flags=re.IGNORECASE)  # replace "U.S." or "US" with "usa"
    
    text = re.sub(r'<[^>]+>', '', text)  # remove HTML tags
    text = re.sub(r'&\w+;', '', text)  # remove HTML entities
    
    text = re.sub(r'[^a-z\s]', '', text)  # remove numbers and punctuation (keep only letters and spaces)
    
    words = text.split()
    words = [word for word in words if word not in stop_words] # remove stop-words, tokenization
    
    words = [lemmatizer.lemmatize(word) for word in words] # lemmatization
    
    return ' '.join(words) # combining words into a string
```

```python
# Applying Preprocessing Function

df_train['clean_title'] = df_train['Title'].apply(preprocess_text)
df_train['clean_description'] = df_train['Description'].apply(preprocess_text)
```

### Exploratory Data Analysis (EDA)

```python
# Data Integrity Check: Missing Values

df_train.isnull().sum()
```

```python
# Determining unique classes and their frequency distribution

df_train['Class Index'].unique() # get unique classes
```

```python
df_train['Class Index'] = df_train['Class Index'].astype('category')  # convert 'Class Index' column to a categorical data type
sns.countplot(x='Class Index', data=df_train)  # create a count plot to show the classes distribution 
plt.show()  # display the plot
```

```python
# Adding word and character count columns

def count_words(text): # function for counting words in a text
    return len(text.split())

def count_characters(text): # function for counting characters in text
    return len(text)

# Create new columns
df_train['word_count_clean_itle'] = df_train['clean_title'].apply(count_words)
df_train['character_count_clean_title'] = df_train['clean_title'].apply(count_characters)
df_train['character_count_clean_description'] = df_train['clean_description'].apply(count_characters)
```

```python
# Distibution of words and characters count

df_train.describe() # summary statistics for numerical columns in the dataset
```

```python
# Distribution of title and descriptions text lengths

## Using tokenizer to correctly split the text into words, taking punctuation into account

df_train['title_text_length'] = df_train['clean_title'].apply(lambda x: len(word_tokenize(x))) # title lengths distibution
sns.histplot(df_train['title_text_length'], bins=50)
plt.show()
```

```python
df_train['description_text_length'] = df_train['clean_description'].apply(lambda x: len(word_tokenize(x))) # description lengths distibution
sns.histplot(df_train['description_text_length'], bins=50)
plt.show()
```

```python
# Testing the Normality of Distributions

# Shapiro-Wilk test
shapiro_title = shapiro(df_train['title_text_length'][:5000])  #    5000
shapiro_description = shapiro(df_train['description_text_length'][:5000])

# D'Agostino's K^2 test (   )
dagostino_title = normaltest(df_train['title_text_length'])
dagostino_description = normaltest(df_train['description_text_length'])

# Tests results display
print("Shapiro-Wilk test for title lengths:", shapiro_title)
print("Shapiro-Wilk test for description lengths:", shapiro_description)
print("D'Agostino's K^2 test for title lengths:", dagostino_title)
print("D'Agostino's K^2 test for description lengths:", dagostino_description)
```

```python
# Word Frequency Analysis

## Titles

### 50 most frequent words in titles

all_words_clean_title = ' '.join(df_train['clean_title']).split() # creating list of individual words
freq_dist_clean_title = Counter(all_words_clean_title) # word count of the words in the titles
print(freq_dist_clean_title.most_common(50)) # print the words 
```

```python
### Visualizing the Top 50 Most Common Words in Titles

most_common_words_clean_title = freq_dist_clean_title.most_common(10) # get the 50 most common words from the frequency distribution of titles

words_clean_title = [word[0] for word in most_common_words_clean_title] # extract words 
counts_clean_title = [word[1] for word in most_common_words_clean_title] # extract words' frequencies

plt.figure(figsize=(15, 2)) # define the size of the plot
sns.barplot(x=counts_clean_title, y=words_clean_title) # create a horizontal bar plot of word frequencies
plt.title('Top 50 Most Common Words in Titles') # add title and labels to the plot
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show() # display the plot
```

```python
### Generating a Word Cloud for most frequent words in Titles

text = " ".join(title for title in df_train.clean_title) # combine all titles into a single string

wordcloud_title = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = STOPWORDS,
                min_font_size = 10).generate(text)

plt.figure(figsize = (8, 8), facecolor = None)   # display the Word Cloud
plt.imshow(wordcloud_title) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show()
```

```python
## Descriptions

all_words_description = ' '.join(df_train['clean_description']).split() # creating list of individual words
freq_dist_description = Counter(all_words_description)      # word count of the words in the descriptions
print(freq_dist_description.most_common(50))          # print 50 most common words in descriptions
```

```python
### Visualizing the Top 50 Most Common Words in Descriptions

most_common_words_clean_description = freq_dist_description.most_common(10) # get the 50 most common words from the frequency distribution of titles

words_clean_description = [word[0] for word in most_common_words_clean_description] # extract words 
counts_clean_description = [word[1] for word in most_common_words_clean_description] # extract words' frequencies

plt.figure(figsize=(15, 2)) # define the size of the plot
sns.barplot(x=counts_clean_description, y=words_clean_description) # create a horizontal bar plot of word frequencies
plt.title('Top 50 Most Common Words in Descriptions') # add title and labels to the plot
plt.xlabel('Frequency')
plt.ylabel('Words')

plt.show() # display the plot
```

```python
### Generating a Word Cloud for most frequent words in Descriptions

text = " ".join(description for description in df_train.clean_description) # combine all descriptions into a single string

wordcloud_description = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = STOPWORDS,
                min_font_size = 10).generate(text)

plt.figure(figsize = (8, 8), facecolor = None)   # display the Word Cloud
plt.imshow(wordcloud_description) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```

```python
# Plot Zipf's Law for Titles: rank vs frequency on a log-log scale

## Using the already prepared data for words and their frequencies

log_freqs_clean_title = np.log(counts_clean_title) # logarithms of frequencies
log_ranks_clean_title = np.log(range(1, len(counts_clean_title) + 1))  # logarithms of ranks

## Plotting Zipf's Law

plt.figure(figsize=(10, 6))
sns.regplot(x=log_ranks_clean_title, y=log_freqs_clean_title)

plt.xlabel('Log of Rank')  # X-axis label
plt.ylabel('Log of Frequency')  # Y-axis label
plt.title('Log-Log Plot with Linear Regression (Using Clean Title Data)')  # title

plt.show() # displaying the plot
```

```python
# Plot Zipf's Law for Descriptions: rank vs frequency on a log-log scale

## Using the already prepared data for words and their frequencies

log_freqs_clean_descriptions = np.log(counts_clean_description) # logarithms of frequencies
log_ranks_clean_description = np.log(range(1, len(counts_clean_description) + 1))  # logarithms of ranks

## Plotting Zipf's Law

plt.figure(figsize=(10, 6))
sns.regplot(x=log_ranks_clean_description, y=log_freqs_clean_descriptions)

plt.xlabel('Log of Rank')  # X-axis label
plt.ylabel('Log of Frequency')  # Y-axis label
plt.title('Log-Log Plot with Linear Regression (Using Clean Description Data)')  # title

plt.show() # displaying the plot
```
